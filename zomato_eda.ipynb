```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('zomato.csv',encoding='latin-1')
```

```python
  dfc =pd.read_excel('Country-Code.xlsx')
```

```python
dfc.head()
```

```python
final_df=pd.merge(df,dfc,on="Country Code",how="left")
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation:Zomato maximum records are from India after that USA and then UK

```python
rating=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns ={0:'Rating Count'})
```

```python
print(rating)
```

Observation

1.when rating is between 4.5 to 4.9 -->Excellent

2.when rating is between 4 to 4.4 -->Vert good
3.when rating is between 3.5 to 3.9 -->Good
4.when rating is between 2.5 to 3.4 -->Average
5.when rating is between 1.8 to 2.4 -->Poor



```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color', data = rating)
```

## observation
1.not rated count is very high
2. Maximum no of rating are b/w 2.5 to 3.4
